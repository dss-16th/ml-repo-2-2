# Scaling
- MinMaxScaler
    - 모든 feature가 0과 1사이에 위치하게 만듭니다.
      데이터가 2차원 셋일 경우, 모든 데이터는 x축의 0과 1 사이에, y축의 0과 1사이에 위치하게 됩니다.
- StandardScaler
    - 각 feature의 평균을 0, 분산을 1로 변경합니다. 
- RobustScaler
    - 모든 특성들이 같은 스케일을 갖게 됩니다.모든 특성들이 같은 크기를 갖는다는 점에서 StandardScaler와 비슷하지만, 평균과 분산 대신 median과 quartile을 사용합니다. RobustScaler는 이상치에 영향을 받지 않습니다.
- Normalizer
    - StandardScaler, RobustScaler, MinMaxScaler가 각 columns의 통계치를 이용한다면 Normalizer는 row마다 각각 정규화됩니다. Normalizer는 유클리드 거리가 1이 되도록 데이터를 조정합니다. (유클리드 거리는 두 점 사이의 거리를 계산할 때 쓰는 방법, L2 Distance)
    

```python
from sklearn.preprocessing import (MinMaxScaler, StandardScaler, RobustScaler, Normalizer)

MMS = MinMaxScaler()
SS = StandardScaler()
RS = RobustScaler()
NM = Normalizer()
```

```python
mms_train = X_resampled.copy()
ss_train = X_resampled.copy()
rs_train = X_resampled.copy()
nm_train = X_resampled.copy()

mms_test = X_test.copy()
ss_test = X_test.copy()
rs_test = X_test.copy()
nm_test = X_test.copy()
```

```python
mms_train[['x14','x15']] = MMS.fit_transform(mms_train[['x14','x15']])
ss_train[['x14','x15']] = SS.fit_transform(ss_train[['x14','x15']])
rs_train[['x14','x15']] = RS.fit_transform(rs_train[['x14','x15']])
nm_train[['x14','x15']] = NM.fit_transform(nm_train[['x14','x15']])

mms_test[['x14','x15']] = MMS.transform(mms_test[['x14','x15']])
ss_test[['x14','x15']] = SS.transform(ss_test[['x14','x15']])
rs_test[['x14','x15']] = RS.transform(rs_test[['x14','x15']])
nm_test[['x14','x15']] = NM.transform(nm_test[['x14','x15']])
```

### modeling

```python
# logistic regression
import time
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced


log_clf = LogisticRegression(C=0.01, class_weight="balanced", solver="liblinear", random_state=4)
# log_clf = LogisticRegression(C=0.01, solver="lbfgs", random_state=4)
# log_clf = LogisticRegression(C=0.001, dual=False, fit_intercept=True,
#                              n_jobs=None, random_state=4, solver='liblinear')
log_clf.fit(mms_train, y_resampled)
log_pred = log_clf.predict(X_test)

# 모델과 성능 출력 함수
def get_result(model, mms_train, y_resampled, mms_test, y_test):
    start_time = time.time()
    model.fit(mms_train, y_resampled)
    pred = model.predict(mms_test)
    
    print("√", model, "\n")
    print(confusion_matrix(y_test, pred), "\n------------------------------------------------------")
    print(classification_report_imbalanced(y_test, pred, digits=6), "\n-----------------------------------------------------------------------------------")
    print('Fit time : ', time.time() - start_time, "\n-----------------------------------------------------------------------------------\n")

get_result(log_clf, mms_train, y_resampled, mms_test, y_test)
get_result(log_clf, ss_train, y_resampled, ss_test, y_test)
get_result(log_clf, rs_train, y_resampled, rs_test, y_test)
get_result(log_clf, nm_train, y_resampled, nm_test, y_test)
```
