### 1. re-labeling
- ÎùºÎ≤®Í≥º ÎÇ¥Ïö©Ïùò ÏàúÏÑúÍ∞Ä Ïù¥Ïñ¥ÏßÄÏßÄ ÏïäÏïÑÏÑú re-labeling ÏßÑÌñâ

```python
relabel
```

```python
df_relabel = socar[relabel]
df_relabel
```

```python
# Î≥ÄÍ≤Ω Ï†Ñ
df_relabel['x6'].value_counts()
```

```python
# Î≥ÄÍ≤Ω ÌõÑ
mapping_value = {2:1}
df_relabel = df_relabel.replace({'x6': mapping_value})
df_relabel['x6'].value_counts()
```

```python
# Î≥ÄÍ≤Ω Ï†Ñ
df_relabel['x12'].value_counts()
```

```python
# Î≥ÄÍ≤Ω ÌõÑ
mapping_value = {5:0, 2:3, 3:4, 4:5}
df_relabel = df_relabel.replace({'x12': mapping_value})
df_relabel['x12'].value_counts()
```

```python
# Î≥ÄÍ≤Ω Ï†Ñ
df_relabel['x21'].value_counts()
```

```python
# Î≥ÄÍ≤Ω ÌõÑ
mapping_value = {2:0}
df_relabel = df_relabel.replace({'x21': mapping_value})
df_relabel['x21'].value_counts()
```

```python
# Î≥ÄÍ≤Ω Ï†Ñ
df_relabel['x13'].value_counts()
```

```python
# Î≥ÄÍ≤Ω ÌõÑ
mapping_value = {-1:0}
df_relabel = df_relabel.replace({'x13': mapping_value})
df_relabel['x13'].value_counts()
```

```python
# Î≥ÄÍ≤Ω Ï†Ñ
df_relabel['x23'].value_counts()
```

```python
# Î≥ÄÍ≤Ω ÌõÑ
mapping_value = {-1:0, 0:1, 1:2, 2:3, 3:4, 4:5}
df_relabel = df_relabel.replace({'x23': mapping_value})
df_relabel['x23'].value_counts()
```

```python
df_relabel
```

```python
socar_relabel = socar.copy()
```

```python
# Îç∞Ïù¥ÌÑ∞ ÍµêÏ≤¥ÌïòÍ∏∞
socar_relabel = socar_relabel.drop(columns=relabel)
socar_relabel = pd.concat((socar_relabel, df_relabel), axis=1)
```

```python
socar_relabel
```

#### train, test Î∂ÑÎ¶¨

```python
train = socar_relabel[socar_relabel['x24']==0]
train.drop(columns=['x24'], inplace=True)
test = socar_relabel[socar_relabel['x24']==1]
test.drop(columns=['x24'], inplace=True)
```

```python
train.reset_index(drop=True, inplace=True)
test.reset_index(drop=True, inplace=True)
```

```python
train
```

```python
test
```

### 2. OneHotEncoder()
- Í∏∞Ï°¥Ïóê ÎùºÎ≤®ÎßÅ ÎêòÏñ¥ÏûàÎäî Î≥ÄÏàòÎì§ Ï§ëÏóê ÏàúÏÑúÍ∞Ä ÏóÜÎäî Î≥ÄÏàòÎäî ÏõêÌï´Ïù∏ÏΩîÎî© ÏßÑÌñâ

```python
# ÏõêÌï´Ïù∏ÏΩîÎî©Ìï† Î≥ÄÏàò
onehot
```

```python
# train
```

```python
df_onehot_train = train[onehot]
df_onehot_train
```

```python
# 33Í∞ú ÏÉùÍ∏¥Îã§
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
enc.fit(df_onehot_train)
df_onehot_train = pd.DataFrame(data=enc.transform(df_onehot_train), columns=enc.get_feature_names(onehot), dtype=bool)
# trasfer true and false to 1 and 0
df_onehot_train = df_onehot_train * 1
```

```python
df_onehot_train
```

```python
df_onehot_train.columns, len(df_onehot_train.columns)
```

```python
# Îç∞Ïù¥ÌÑ∞ Ìï©ÏπòÍ∏∞
train_encode = pd.concat((train, df_onehot_train), axis=1)
train_encode = train_encode.drop(columns=onehot)
```

```python
train_encode.columns, len(train_encode.columns)
```

```python
# test
```

```python
df_onehot_test = test[onehot]
df_onehot_test
```

```python
df_onehot_test = pd.DataFrame(data=enc.transform(df_onehot_test), columns=enc.get_feature_names(onehot), dtype=bool)
df_onehot_test = df_onehot_test * 1
```

```python
df_onehot_test
```

```python
# Îç∞Ïù¥ÌÑ∞ Ìï©ÏπòÍ∏∞
test_encode = pd.concat((test, df_onehot_test), axis=1)
test_encode = test_encode.drop(columns=onehot)
```

```python
test_encode.columns, len(test_encode.columns)
```

```python
# re-label, OneHot ÏôÑÎ£å
train_encode.to_csv('datas/train_encode_ver2.csv')
test_encode.to_csv('datas/test_encode_ver2.csv')
```
